## Tratamientos de Datos Faltantes

```python
dataset[dataset['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (dataset['Mantenimiento'].isnull()) & (dataset['Tipo'] == 'Departamento')
seleccion
```

```python
A = dataset.shape[0]
dataset = dataset[~seleccion]
B = dataset.shape[0]
A - B
```

```python
dataset[dataset['Mantenimiento'].isnull()].shape[0]
```

```python
dataset = dataset.fillna({'Mantenimiento':0, 'Impuesto':0})
```


## Creando Nuevas Variables

### Valor *Bruto*


```python
dataset['Valor Bruto'] = (dataset['Valor']) + (dataset['Mantenimiento']) + (dataset['Impuesto'])
dataset.head(10)
```

```python
dataset['Valor m2'] = (dataset['Valor']) / (dataset['Area'])
dataset.head(10)
```

```python
dataset['Valor m2'] = dataset['Valor m2'].round(2)
dataset.head(10)
```

```python
dataset['Valor Bruto m2'] = ((dataset['Valor Bruto']) / (dataset['Area'])).round(2)
dataset.head(10)
```

```python
casa = ['Casa', 'Casa en condominio', 'Casa de villa']
```

```python
dataset['Tipo Agrupado'] = dataset['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Departamento')
```

```python
dataset
```
