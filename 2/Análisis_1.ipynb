## Test de hipótesis

```python
sns.lmplot(x='iu', y='valor_factura', data=tarjetas, palette='inferno', col='moroso');
```

```python
from scipy.stats import ranksums
```

### Hipótesis
La distribución de los grupos moroso y no moroso es la misma

### Hipótesis: Alternativa
La distribución de los grupos moroso y no moroso NO es la misma

```python
moroso = tarjetas.query('moroso == 1').valor_factura
moroso
```

```python
no_moroso = tarjetas.query('moroso == 0').valor_factura
no_moroso
```

```python
resultado = ranksums(moroso, no_moroso)
```

```python
print('El p_value del test es {}'. format(resultado.pvalue))
```

La distribución del valor de factura entre los grupos moroso y no moroso no es igual, por lo que podemos aceptar la hipótesis alternativa.
